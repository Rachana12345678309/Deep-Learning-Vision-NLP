# üß™ CNN Hands-On: CIFAR-10 Image Classification

In this exercise, you will work with the **CIFAR-10 dataset**, a popular benchmark dataset used in computer vision.

---

## üì¶ About the Dataset

- Contains **60,000 color images**
- Each image size: **32√ó32 pixels**
- **10 categories**:

`Airplane, Automobile, Bird, Cat, Deer, Dog, Frog, Horse, Ship, Truck`

- Split into:
  - **50,000 training images**
  - **10,000 testing images**

---

## üéØ Your Task

Build a **Convolutional Neural Network (CNN)** that can classify images into one of the **10 classes**.

You will:

1. Load the CIFAR-10 dataset  
2. Normalize the images  
3. Build a simple CNN  
4. Train and evaluate the model  
5. Test predictions on sample images

```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import numpy as np

(x_train, y_train), (x_test, y_test) = datasets.cifar10.load_data()
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

maxlen = 100
x_train = pad_sequences(x_train,maxlen = maxlen)
x_test = pad_sequences(x_test,maxlen = maxlen)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Dense, SimpleRNN, LSTM
from tensorflow.keras.preprocessing.text import Tokenizer


num_words = 10000

model = Sequential()
model.add(Embedding(input_dim = num_words, output_dim = 32, input_length = maxlen))
model.add(SimpleRNN(units = 64, return_sequences = False)) #one output only
model.add(Dense(1, activation = 'sigmoid')) # 1 since binary classification

model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(x_train, y_train, epochs = 10, batch_size= 64, validation_split = 0.2)
```
